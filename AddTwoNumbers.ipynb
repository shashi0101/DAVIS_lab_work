```python
# open file in a read mode
filev = open("firstfile.txt", "r")

data = filev.read()

count = 0
# checking each character
for ch in data:
    if ch in 'aeiouAEIOU':   # both cases included
        count += 1

print("Number of vowels in file:", count)
# close file
filev.close()
```
