```python
# To calculate area and perimeter of circle
#------------------------------------------------
# take user input
r = float(input("Enter radius: "))
# formula
area = 3.14 * r * r
perimeter = 2 * 3.14 * r
# print the result
print("Area =", area)
print("Perimeter =", perimeter)
```
