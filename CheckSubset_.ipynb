```python
# Define two sets
set_a = {1, 2}
set_b = {1, 2, 3, 4}

# Check if set_a is a subset of set_b using the <= operator
is_subset = set_a <= set_b

print(f"Is set_a a subset of set_b? {is_subset}") # Output: True
```
