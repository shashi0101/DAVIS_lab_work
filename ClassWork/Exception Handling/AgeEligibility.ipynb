```python
try:
  age=float(input("Enter age:"))
  #---------------------------------------
  print("----------------------------")
  if(age>18):
    print("Eligible for Voting")
  elif(age<0):
    print("Enter valid Age")
  else:
    print("not Eligible for Voting")
except ValueError:
  print("enter a valid number")
else:
  print("----------------------------")
finally:
  print("---------Program Executed--------")

```
