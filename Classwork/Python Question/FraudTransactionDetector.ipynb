```python
# Fraud Transaction Detector

transactions = [
    # user, amount, time
    ("A", 5000, 1),
    ("B", 2000, 2),
    ("A", 7000, 3),
    ("A", 3000, 4),
    ("B", 9000, 5)
]

threshold = 6000   # amount limit
flagged = {}       # dictionary to store fraud

last_time = {}     # store last transaction time of each user

for t in transactions:
    user = t[0]
    amount = t[1]
    time = t[2]

    # condition 1: amount greater than threshold
    if amount > threshold:
        flagged[user] = "High Amount"

    # condition 2: multiple transactions in short time (difference <= 2)
    if user in last_time:
        if time - last_time[user] <= 2:
            flagged[user] = "Too Many Transactions"

    # update last transaction time
    last_time[user] = time

# print fraud transactions
print("Fraud Transactions:")
for user in flagged:
    print(user, "->", flagged[user])
```
