```python
# Smart Student Result Analyzer

students = []   # list to store data

# open file
try:
    f = open("students.txt", "r")

    for line in f:
        try:
            data = line.strip().split(",")

            # check if data is complete
            if len(data) < 5:
                print("Incomplete data:", line.strip())
                continue

            # take values
            sid = int(data[0])
            name = data[1]
            m1 = int(data[2])
            m2 = int(data[3])
            m3 = int(data[4])

            # calculate total and percentage
            total = m1 + m2 + m3
            per = total / 3

            # grade using if-else
            if per >= 90:
                grade = "A+"
            elif per >= 75:
                grade = "A"
            elif per >= 60:
                grade = "B"
            elif per >= 50:
                grade = "C"
            else:
                grade = "Fail"

            # store data
            students.append([sid, name, total, per, grade])

        except:
            print("Invalid data:", line.strip())

    f.close()

except:
    print("File not found!")

# display result
print("\nStudent Results")
for s in students:
    print(s[0], s[1], "Total:", s[2], "Per:", round(s[3],2), "Grade:", s[4])

# find topper
topper = students[0]

for s in students:
    if s[3] > topper[3]:
        topper = s

# display topper
print("\nTopper:")
print(topper[0], topper[1], "Per:", round(topper[3],2))
```
