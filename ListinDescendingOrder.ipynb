```python
a = []
x = int(input("Enter the number of elements: "))
for i in range(x):
    element = input(f"Enter element {i+1}: ")
    a.append(element)

print("List:", a)
n = len(a)

for i in range(n):
    for j in range(0, n - i - 1):
        if a[j] < a[j + 1]:  # Descending
            a[j], a[j + 1] = a[j + 1], a[j]

print("sorted array in descending order:",a)
```
