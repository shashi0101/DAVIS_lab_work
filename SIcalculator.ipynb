```python
#function to calculate simple interest
def calculateSI (p,r,t):
  return((p*r*t)/100)

  #main program

  principal=float(input("enter principal(in Rs):"))
  rate=float(input("enter rate of interst(in Rs):"))
  time=int(input("enter time (in year):"))
  print("simple interest is:",calculateSI(principal,rate,time))
```
