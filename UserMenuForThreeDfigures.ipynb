```python
import geometry

def main_menu():
    print("\n--- 3D Figure Calculator ---")
    print("1. Cube\n2. Cuboid\n3. Cylinder\n4. Cone\n5. Sphere\n6. Hemisphere\n0. Exit")
    choice = input("Select a figure: ")

    if choice == '0': return False

    print("\nSelect Operation:\n1. Curved Surface Area (CSA)\n2. Total Surface Area (TSA)\n3. Volume")
    op = input("Select operation: ")

    if choice == '1': # Cube
        s = float(input("Enter side: "))
        if op == '1': print("Result:", geometry.cube_csa(s))
        elif op == '2': print("Result:", geometry.cube_tsa(s))
        elif op == '3': print("Result:", geometry.cube_vol(s))

    elif choice == '2': # Cuboid
        l = float(input("Enter length: ")); w = float(input("Enter width: ")); h = float(input("Enter height: "))
        if op == '1': print("Result:", geometry.cuboid_csa(l, w, h))
        elif op == '2': print("Result:", geometry.cuboid_tsa(l, w, h))
        elif op == '3': print("Result:", geometry.cuboid_vol(l, w, h))

    elif choice in ['3', '4']: # Cylinder or Cone
        r = float(input("Enter radius: ")); h = float(input("Enter height: "))
        if choice == '3':
            if op == '1': print("Result:", geometry.cylinder_csa(r, h))
            elif op == '2': print("Result:", geometry.cylinder_tsa(r, h))
            elif op == '3': print("Result:", geometry.cylinder_vol(r, h))
        else:
            if op == '1': print("Result:", geometry.cone_csa(r, h))
            elif op == '2': print("Result:", geometry.cone_tsa(r, h))
            elif op == '3': print("Result:", geometry.cone_vol(r, h))

    elif choice in ['5', '6']: # Sphere or Hemisphere
        r = float(input("Enter radius: "))
        if choice == '5':
            if op == '1': print("Result:", geometry.sphere_csa(r))
            elif op == '2': print("Result:", geometry.sphere_tsa(r))
            elif op == '3': print("Result:", geometry.sphere_vol(r))
        else:
            if op == '1': print("Result:", geometry.hemi_csa(r))
            elif op == '2': print("Result:", geometry.hemi_tsa(r))
            elif op == '3': print("Result:", geometry.hemi_vol(r))
    return True

running = True
while running:
    running = main_menu()
```
