```python
# program for writing data into files
# opening file for write operation
filev = open("firstfile.txt", "w")
# blank list of sentences
sentences = []
# writing 20 sentence into the file
print ("Enter any 20 sentences:")
for x in range(20):
    # input of data from user
    sentence = input()
    # inserting sentences in to the list
    sentences.append(sentence)
    print("---------------------------------------")
    # writing data into file
    filev.writelines(sentence)
print("Data seccesfully written")
# closing the file
filev.close()
```
