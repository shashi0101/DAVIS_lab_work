```python
#function for classifying age group
def classify_age_group(age):
    if age < 18:
        return "Minor"
    elif 18 <= age < 65:
        return "Adult"
    else:
        return "Senior"

# main function
input_age_1 = 65
output_group_1 = classify_age_group(input_age_1)
print(f"Input: {input_age_1}")
print(f"Output: {output_group_1}")

```
