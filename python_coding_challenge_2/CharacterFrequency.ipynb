```python
#count occurences of a character
def count_char_occurrences(input_string, char_to_count):

    #Counts the occurrences of a specific character in a string.


    count = 0
    for char in input_string:
        if char == char_to_count:
            count += 1
    return count

# main function
input_str = "banana"
input_char = 'a'
output = count_char_occurrences(input_str, input_char)
print(f"Input: \"{input_str}\", '{input_char}'")
print(f"Output: {output}")

```
