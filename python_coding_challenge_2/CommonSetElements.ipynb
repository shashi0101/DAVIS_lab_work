```python
def find_set_intersection(set1, set2):
    # Use the intersection() method to find the intersection of the two sets
    return set1.intersection(set2)
    # Alternatively, you can use the & operator:
    # return set1 & set2

# main finction
input_set1_1 = {1, 2}
input_set1_2 = {2, 3}
output_1 = find_set_intersection(input_set1_1, input_set1_2)
print(f"Input sets: {input_set1_1}, {input_set1_2}")
print(f"Intersection: {output_1}")

```
