```python
def count_digits(number):

    #Counts the number of digits in a given integer.


    # Convert the number to a string and return its length
    # Use absolute value to handle negative numbers correctly (e.g., -123 has 3 digits)
    return len(str(abs(number)))

# main function
input_num = 45678
output_count_digits = count_digits(input_num)
print(f"Input: {input_num}")
print(f"Output: {output_count_digits}")

```
