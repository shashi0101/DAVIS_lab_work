```python
#calculate high value entries
def count_numbers_greater_than_50(numbers):

    #Counts the number of entries in a list that are greater than 50.


    count = 0
    for num in numbers:
        if num > 50:
            count += 1
    return count

# main function
input_list_1 = [10, 60, 30, 80]
output_count_1 = count_numbers_greater_than_50(input_list_1)
print(f"Input: {input_list_1}")
print(f"Output: {output_count_1}")

```
