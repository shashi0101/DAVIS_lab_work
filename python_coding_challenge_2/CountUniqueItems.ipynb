```python
def count_unique_elements(input_list):
    # Convert the list to a set to get unique elements, then return its length
    return len(set(input_list))

# main function
input = [1, 1, 2, 3, 3]
output = count_unique_elements(input)
print(f"Input: {input}")
print(f"Number of unique elements: {output}")

```
