```python
def count_words(sentence):
    # Split the sentence by whitespace to get a list of words
    words = sentence.split()
    # Return the number of words in the list
    return len(words)

# main function
input_str = "Python is great"
output = count_words(input_str)
print(f"Input: '{input_str}'")
print(f"Number of words: {output}")

```
