```python
from collections import Counter

def find_duplicate_chars(input_string):
    # Use Counter to get the frequency of each character
    char_counts = Counter(input_string)

    duplicate_chars = []
    # Iterate through the characters and their counts
    for char, count in char_counts.items():
        if count > 1:
            duplicate_chars.append(char)

    # Sort the characters for consistent output, then join them with spaces
    return " ".join(sorted(duplicate_chars))

# Example usage:
input = "programming"
output = find_duplicate_chars(input)
print(f"Input: '{input}'")
print(f"Output: '{output}'")

```
