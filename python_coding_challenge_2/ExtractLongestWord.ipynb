```python
def find_longest_word(sentence):
    # Split the sentence into words
    words = sentence.split()

    # Handle empty sentence case
    if not words:
        return ""

    # Initialize longest_word with the first word
    longest_word = words[0]

    # Iterate through the words to find the longest one
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word

    return longest_word

# main function
input_str = "I love programming"
output_word = find_longest_word(input_str)
print(f"Input: '{input_str}'")
print(f"Longest word: '{output_word}'")

```
