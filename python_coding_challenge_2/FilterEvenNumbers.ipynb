```python
#function to calculate even numbers
def print_even_numbers(N):

   # Prints all even numbers from 1 to N.


    even_numbers = []
    for i in range(1, N + 1):
        if i % 2 == 0:
            even_numbers.append(str(i))
    print(" ".join(even_numbers))

# main function
input_N_1 = 10
print(f"Input: {input_N_1}")
print(f"Output:", end=" ")
print_even_numbers(input_N_1)

```
