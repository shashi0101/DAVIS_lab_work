```python
def find_common_elements(list1, list2):
    # Convert lists to sets to efficiently find common elements
    set1 = set(list1)
    set2 = set(list2)

    # Find the intersection of the two sets
    common_elements = set1.intersection(set2)

    # Convert the set of common elements back to a list
    return list(common_elements)

# Example usage:
input1 = [1, 2, 3]
input2 = [2, 3, 4]
output_list = find_common_elements(input1, input2)
print(f"Common elements between {input1} and {input2}: {output_list}")

```
