```python
def find_max_without_builtin(input_list):

    #Finds the maximum value in a list of numbers without using built-in functions like max().


    if not input_list:
        return None # Handle empty list case

    # Assume the first element is the maximum initially
    maximum = input_list[0]

    # Iterate through the rest of the list, comparing each element
    for item in input_list[1:]:
        if item > maximum:
            maximum = item
    return maximum

# main function
input_list_1 = [5, 8, 2]
output_max_1 = find_max_without_builtin(input_list_1)
print(f"Input: {input_list_1}")
print(f"Output: {output_max_1}")

```
