```python
def find_second_largest(numbers):
    # Remove duplicates and sort the list in descending order
    unique_numbers = sorted(list(set(numbers)), reverse=True)

    # Check if there are at least two unique elements
    if len(unique_numbers) < 2:
        return "Cannot find second largest element (less than two unique elements)"
    else:
        return unique_numbers[1]

# main function
input = [10, 20, 5, 15]
output = find_second_largest(input)
print(f"Input: {input}")
print(f"Second largest element: {output}")

```
