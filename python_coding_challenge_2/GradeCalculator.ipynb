```python
#calculateGrade
def assign_grade(marks):

    #using if else to assign a letter grade based on the given marks.

      if 90 <= marks <= 100:
        return "A"
      elif 80 <= marks < 90:
        return "B"
      elif 70 <= marks < 80:
        return "C"
      elif 60 <= marks < 70:
        return "D"
      else:
        return "F"

# main function
marks = 82
output = assign_grade(marks)
print(f"Input: {marks}")
print(f"Output: {output}")

```
