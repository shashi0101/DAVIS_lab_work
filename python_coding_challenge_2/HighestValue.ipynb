```python
#Calculate largest number
def find_largest_number(num1, num2, num3):
    # Method 1: Using conditional statements
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3


# main function
num1 = 10
num2 = 25
num3 = 15
output = find_largest_number(num1, num2, num3)
print(f"Input: {num1}, {num2}, {num3}")
print(f"Output: {output}")

```
