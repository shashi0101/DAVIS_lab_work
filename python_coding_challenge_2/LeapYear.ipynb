```python
#checking leap year
def is_leap_year(year):
  #using if else to check the condition
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return "Leap Year"
    else:
        return "Not a Leap Year"

# main function
input_year_1 = 2024
output_1 = is_leap_year(input_year_1)
print(f"Input: {input_year_1}")
print(f"Output: {output_1}")

```
