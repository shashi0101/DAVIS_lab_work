```python
def find_set_union(set1, set2):
    # Use the union() method to find the union of the two sets
    return set1.union(set2)
    # Alternatively, you can use the | operator:
    # return set1 | set2

# Example usage:
input_set1_1 = {1, 2}
input_set1_2 = {2, 3}
output_1 = find_set_union(input_set1_1, input_set1_2)
print(f"Input sets: {input_set1_1}, {input_set1_2}")
print(f"Union: {output_1}")

```
