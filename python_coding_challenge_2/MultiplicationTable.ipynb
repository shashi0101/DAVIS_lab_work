```python
#function to calculate multiplication
def print_multiplication_table(number):

    #Prints the multiplication table for a given number up to 10.


    results = []
    for i in range(1, 11):
        results.append(str(number * i))
    print(" ".join(results))

# main function
input_num = 3
print(f"Input: {input_num}")
print(f"Output:", end=" ")
print_multiplication_table(input_num)
```
