```python
# calculatedivisibility
def check_divisibility_by_3_and_5(number):

    if number % 3 == 0 and number % 5 == 0:
        return "Yes"
    else:
        return "No"

# Main function
num = 15
output = check_divisibility_by_3_and_5(num)
print(f"Input: {num}")
print(f"Output: {output}")

```
