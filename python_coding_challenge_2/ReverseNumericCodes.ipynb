```python
#calculates reverse
def reverse_number(n):

   # Reverses the digits of an integer.


    # Handle negative numbers by reversing the absolute value and reapplying the sign
    sign = -1 if n < 0 else 1
    n = abs(n)

    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
    return reversed_num * sign

# main function
input_num_1 = 1234
output_num_1 = reverse_number(input_num_1)
print(f"Input: {input_num_1}")
print(f"Output: {output_num_1}")

```
