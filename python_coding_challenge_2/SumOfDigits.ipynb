```python
#calculate sum of digits
def sum_digits(number):

    #Calculates the sum of the digits of a given integer.


    s = 0
    # Convert the number to its absolute value to handle negative inputs gracefully
    # Although for 'sum of digits' positive numbers are usually implied.
    abs_number = abs(number)
    for digit in str(abs_number):
        s += int(digit)
    return s

# main function
input_num = 123
output_sum_digits = sum_digits(input_num)
print(f"Input: {input_num}")
print(f"Output: {output_sum_digits}")

```
