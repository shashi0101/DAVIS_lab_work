```python
def to_uppercase_manual(input_string):
    uppercase_chars = []
    for char in input_string:
        # Check if the character is a lowercase letter (ASCII 'a' to 'z')
        if 'a' <= char <= 'z':
            # Convert to uppercase by subtracting 32 from its ASCII value
            uppercase_chars.append(chr(ord(char) - 32))
        else:
            # Keep other characters as they are
            uppercase_chars.append(char)
    return "".join(uppercase_chars)

# main function
input_str = "abc"
output_str = to_uppercase_manual(input_str)
print(f"Original string: '{input_str}'")
print(f"Uppercase string: '{output_str}'")

```
