```python
#calculates ticket price
def get_ticket_price(day):

    #using if else to determine the ticket price based on the day of the week.


    day_lower = day.lower()

    if day_lower in ['saturday', 'sunday']:
        return 200
    elif day_lower in ['monday', 'tuesday', 'wednesday', 'thursday', 'friday']:
        return 150
    else:
        print(f"Error: Invalid day '{day}'. Please enter a valid day of the week.")
        return None

# Main function
day = "Sunday"
price = get_ticket_price(day)
print(f"Input: {day}")
print(f"Output: {price}")

```
