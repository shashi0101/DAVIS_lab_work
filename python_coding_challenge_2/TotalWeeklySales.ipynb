```python
#calculate total weekly sales
def calculate_weekly_sales(daily_sales):

    #Calculates the total weekly sales from a list of daily sales.


    total_sales = sum(daily_sales)
    return total_sales

# Main function
input_sales_1 = [100, 200, 150, 300, 250, 400, 100]
output_sales_1 = calculate_weekly_sales(input_sales_1)
print(f"Input daily sales: {input_sales_1}")
print(f"Total weekly sales: {output_sales_1}")

```
