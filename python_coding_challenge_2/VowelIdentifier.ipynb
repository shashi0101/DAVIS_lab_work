```python
#function to identify vowel
def is_vowel(char):

    # using if else to checks if a given character is a vowel.


    vowels = "aeiouAEIOU"
    if char in vowels:
        return "Vowel"
    else:
        return "Consonant"

# main function
input_char_1 = 'e'
output_1 = is_vowel(input_char_1)
print(f"Input: {input_char_1}")
print(f"Output: {output_1}")

```
