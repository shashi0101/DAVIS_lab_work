```python
#function to calculate bonus
def calculate_bonus(salary):

    if salary <= 30000:
        bonus_rate = 0.10  # 10% bonus
    elif 30001 <= salary <= 60000:
        bonus_rate = 0.07  # 7% bonus
    else:
        bonus_rate = 0.05  # 5% bonus

    return (salary * bonus_rate)

# main function from where bonus function is called
input_salary = 45000
bonus_amount = calculate_bonus(input_salary)
print(f"Input: {input_salary}")
print(f"Output: Bonus = {bonus_amount}")

```
