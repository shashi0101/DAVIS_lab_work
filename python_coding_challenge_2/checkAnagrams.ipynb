```python
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase
    s1_processed = str1.replace(" ", "").lower()
    s2_processed = str2.replace(" ", "").lower()

    # Check if lengths are different after processing
    if len(s1_processed) != len(s2_processed):
        return False

    # Sort characters and compare
    return sorted(s1_processed) == sorted(s2_processed)

# main function
input_str1 = "listen"
input_str2 = "silent"
output_bool = are_anagrams(input_str1, input_str2)
print(f"Are '{input_str1}' and '{input_str2}' anagrams? {output_bool}")

```
