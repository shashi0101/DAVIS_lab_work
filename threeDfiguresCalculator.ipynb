```python
# geometry.py
PI = 3.14159

# Cube Functions
def cube_csa(s): return 4 * s**2
def cube_tsa(s): return 6 * s**2
def cube_vol(s): return s**3

# Cuboid Functions
def cuboid_csa(l, w, h): return 2 * h * (l + w)
def cuboid_tsa(l, w, h): return 2 * (l*w + w*h + h*l)
def cuboid_vol(l, w, h): return l * w * h

# Cylinder Functions
def cylinder_csa(r, h): return 2 * PI * r * h
def cylinder_tsa(r, h): return 2 * PI * r * (r + h)
def cylinder_vol(r, h): return PI * r**2 * h

# Cone Functions (Slant height 'l' = sqrt(r^2 + h^2) represented as (r**2 + h**2)**0.5)
def cone_csa(r, h):
    slant = (r**2 + h**2)**0.5
    return PI * r * slant
def cone_tsa(r, h):
    slant = (r**2 + h**2)**0.5
    return PI * r * (r + slant)
def cone_vol(r, h): return (1/3) * PI * r**2 * h

# Sphere Functions
def sphere_csa(r): return 4 * PI * r**2  # For sphere, CSA = TSA
def sphere_tsa(r): return 4 * PI * r**2
def sphere_vol(r): return (4/3) * PI * r**3

# Hemisphere Functions
def hemi_csa(r): return 2 * PI * r**2
def hemi_tsa(r): return 3 * PI * r**2
def hemi_vol(r): return (2/3) * PI * r**3
```
