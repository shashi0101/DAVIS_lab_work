```python
# program for writing data into files
# opening file for write operation
filev = open("firstfile.txt", "w")
# writing five sentence into the file
print ("Enter any five sentences:")
for x in range(5):
    # input of data from user
    sentence = input()
    # writing sentence into the file
    filev.write(sentence)
    print("---------------------------------------")
print("Data seccesfully written")
# closing the file
filev.close()
```
